```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2
```

```python
xlist = np.linspace(0.1,10,num=1000)
ylist = f(xlist)
```

```python
plt.figure(num=0 , dpi=120)
plt.plot(xlist,ylist, label='linear')
plt.plot(xlist,ylist/2,label="f(x)/2")
plt.title("Ising model")
plt.xlabel("T (temperature)")
plt.ylabel("")
plt.legend()
```
