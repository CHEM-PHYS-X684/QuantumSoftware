**1. Create an zero `BitString` of length 8 and flip a few bits and print the output.**

Methods needed:
- `__str__()`
- `flip()`
- `__len__()`

```python
my_bs = BitString(8)
my_bs.flip_site(2)
my_bs.flip_site(2)
print(" The following should be 0:")
print(my_bs)

my_bs.flip_site(2)
my_bs.flip_site(7)
my_bs.flip_site(0)
print(" The following should have 0,2,7 bits flipped:")
print(my_bs)

print(" Length of bitstring: ", len(my_bs))
assert(len(my_bs) == 8)
```
